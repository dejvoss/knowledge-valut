# Matematyka dla inżynierów w pythonie
## Część 1: Wektory

```python
import matplotlib.pyplot as plt
import math
```

```python
class Punkt:
    def __init__(self, wspolrzedne: list):
        self.point = wspolrzedne
        
    def __str__(self):
        text = [str(p) for p in self.point]
        return "(" + ", ".join(text) + ")"
        
    def plot_point(self):
        fig = plt.figure(figsize=(5,5))
        plt.scatter([self.point[0]], [self.point[1]], label="A", color="black")
        plt.show()
        

class Wektor:
    # jak zaimplementować wektor wielowymiarowy?
    # najpierw musi być punkt wielowymiarowy
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __add__(self, other):
        return Wektor(self.x + other.x, self.y + other.y)
    
    def get_from_points(self, p1, p2):
        self.x = p2.x - p1.x
        self.y = p2.y - p1.y
        
    def __sub__(self, other):
        return Wektor(self.x - other.x, self.y - other.y)
        
    def __mul__(self, other):
        if isinstance(other, int):
            return Wektor(self.x * int, self.y * int)
        elif isinstance(other, Wektor):
            # cos_alpha = other.length() / self.length()
            # return self.length() * other.length() * cos_alpha
            return self.x * other.x + self.y * other.y
            
    def __str__(self):
        return f"[{self.x}, {self.y}]"
    
    def add(self, other):
        colors = ['#00BFFF', '#FFA500', '#32CD32']
        fig = plt.figure(figsize=(5, 5))
        # a
        plt.plot([0, self.x], [0, self.y], label="a", color=colors[0])
        plt.annotate("", xy=(self.x, self.y), xytext=(0, 0), arrowprops=dict(arrowstyle="->", lw=1.5, color=colors[0]))
        # b
        plt.plot([self.x, self.x + other.x], [self.y, self.y + other.y], label="b", color=colors[1])
        plt.annotate("", xy=(self.x + other.x, self.y + other.y), xytext=(self.x, self.y), arrowprops=dict(arrowstyle="->", lw=1.5, color=colors[1]))
        # c = a + b
        plt.plot([0, self.x + other.x], [0, self.y + other.y], label="c=a+b", color=colors[2])
        plt.annotate("", xy=(self.x+other.x, self.y+other.y), xytext=(0,0), arrowprops=dict(arrowstyle="->", lw=1.5, color=colors[2]))
        plt.legend()
        plt.show()
    
    def draw(self):
        fig = plt.figure(figsize=(5, 5))
        plt.plot([0, self.x], [0, self.y])
        plt.annotate(
            "", xy=(self.x, self.y), xytext=(0, 0), arrowprops=dict(arrowstyle="->", lw=1.5)
        )
        plt.show()
        
    def length(self):
        a = self.x
        b = self.y
        return math.sqrt(a**2 + b**2)
    

    
```

```python
p1 = Punkt([4,5])
print(p1)
p1.plot_point()
```

```python
w1 = Wektor(1,3)
print(w1)
w1.draw()
print(w1.length())
```

```python
w2 = Wektor(1,1)
p1 = Punkt(1,2)
p2 = Punkt(3,4)
w2.get_from_points(p1, p2)
print(w2)
print(w2.length())
```

```python
w1.add(w2)
```

```python
print(w1+w2)
```
